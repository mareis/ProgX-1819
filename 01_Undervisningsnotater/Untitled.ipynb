```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mm = plt.imread('mm.jpg')
plt.figure(figsize=(14,12))
plt.imshow(mm)
plt.show()
```

```python
mm=np.array(mm)
```

```python
mm.shape
```

```python
mmd = np.array(mm)
```

```python
for y in range(mm.shape[0]-1):
    for x in range(mm.shape[1]-1):
        mmd[y][x] = mm[y][x+1] - mm[y][x]
```

```python
plt.figure(figsize=(14,12))
plt.imshow(mmd)
plt.show()
```

```python
trafikk = plt.imread('Trafikkskilt.jpg')
trafikk=np.array(trafikk)
trafikkd = np.array(trafikk)
```

```python
plt.figure(figsize=(14,12))
plt.imshow(trafikk)
plt.show()
```

```python
import numpy as np
import scipy
import scipy.misc
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
med_denoised = ndimage.median_filter(trafikk, 3)
```

```python
plt.figure(figsize=(14,12))
plt.imshow(med_denoised)
plt.show()
```

```python
gauss_denoised = ndimage.gaussian_filter(trafikk, 2)
plt.figure(figsize=(14,12))
plt.imshow(gauss_denoised)
plt.show()
```

```python
def bildeDer(bilde):
    bildeD = np.array(bilde)
    for y in range(bilde.shape[0]-1):
        for x in range(bilde.shape[1]-1):
            bildeD[y][x] = bilde[y][x+1] - bilde[y][x]
    return bildeD
```

```python
plt.figure(figsize=(14,12))
plt.imshow(bildeDer(med_denoised))
plt.show()
```
