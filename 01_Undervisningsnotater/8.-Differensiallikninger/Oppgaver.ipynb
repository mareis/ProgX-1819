1)

$y^\prime=1,\; y(0)=0$


```python
y_0 = 0   #Startbetingelse
t_0 = 0   #Første tid
T   = 5   #Tid etter t_0
h   = 0.1 #Steglengde
N   = round(T/h) #Antall steg

import numpy as np
y = np.zeros(N+1) # antall
t = np.linspace(t_0, T, N+1) # Fra , til , antall

y[0] = y_0

def f(y_k, k):
    return 1

for k in range(N):
    y[k+1] = y[k] + h * f(y[k], k)
    
import matplotlib.pyplot as plt

plt.plot(t,y)
plt.show()
```

```python
y_0 = 0   #Startbetingelse
t_0 = 0   #Første tid
T   = 5   #Tid etter t_0
h   = 0.1 #Steglengde
N   = round(T/h) #Antall steg

import numpy as np
y = np.zeros(N+1) # antall
t = np.linspace(t_0, T, N+1) # Fra , til , antall

y[0] = y_0

def f(y_k, t):
    return k

for k in range(N):
    y[k+1] = y[k] + h * f(y[k], k)
    
import matplotlib.pyplot as plt

plt.plot(t,y)
plt.show()
```

3)

$y^\prime(t)=4+3\cdot y(t),\; y(0)=0$

```python
y_0 = 0   #Startbetingelse
t_0 = 0   #Første tid
T   = 5   #Tid etter t_0
h   = 0.1 #Steglengde
N   = round(T/h) #Antall steg

import numpy as np
y = np.zeros(N+1) # antall
t = np.linspace(t_0, T, N+1) # Fra , til , antall

y[0] = y_0

def f(y_k, t):
    return 4 + 3*y_k

for k in range(N):
    y[k+1] = y[k] + h * f(y[k], k)
    
import matplotlib.pyplot as plt

plt.plot(t,y)
plt.show()
```

For å forenkle notasjonen skriver man ofte $y(t)=y$ og $y^\prime(t)=y^\prime$
