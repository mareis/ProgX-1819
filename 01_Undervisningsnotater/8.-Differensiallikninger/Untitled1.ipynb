# Differensialligninger

En diffligning er en ligning på formen

$$x^\prime(t)=f(x(t),t)$$
$$x(a)=x_0$$

der $f$ er en gitt funksjon og $x(a)=x_0$ en startbetingelse. Vi ønsker å finne $x(t)$ for alle $t$



Eksempel: $x^\prime(t)=a*x(t)$ der $a\in\mathbb{R}$ sier at $x(t)$ vokser ($x\prime(t)$) proporsjonalt med seg selv ($ax(t)$) (f.eks kroner på konto, antall individer i en populasjon av dyr/bakterier/osv). 


En mer nyansert modell for populasjon er:

$$x^\prime(t)=x(t)\cdot(a-x(t))$$

$a$ kan ses på som kapasiteten til økosystemet

<a href='https://www.geogebra.org/m/q9csTDeZ'>GeoGebra</a>

Vi skal se på:

$$x^\prime(t)=f(x(t),t)$$
$$x(a)=x_0$$

La $h>0$ være et lite tall og sett $t_k=a+k\cdot h, k=0,1,2,...$

Vi vil finne en tilnærming $x_k\approx x(t_k)$

Vi vet at $f(x(t),t)=x^\prime(t)\approx \frac{x(t+h)-x(t)}{h}$

Sett $t=t_k:$

$f(x(t_k),t_k)\approx \frac{x(t_k+h)-x(t_k)}{h}=\frac{x(t_{k+1})-x(t_k)}{h}$

Erstatt $x(t_k)$ med $x_k$

$$f(x_k, t_k) = \frac{x_{k+1}-x_k}{h}$$

eller

$$x_{k+1} = x_k + h\cdot f(x_k, t_k)$$

```python
import numpy as np
import matplotlib.pyplot as plt
a = -10
x_0 = 1
h = 2**(-7)
T = 1
N = round(T/h)
t = np.linspace(0, T, N+1)
x = np.zeros(N+1)
x[0] = x_0

for k in range(N):
    x[k+1] = x[k] + h*a*x[k]
    
plt.plot(t, x, '-o')
```

## Oppgave 8.2

```python
def diff(f, x_0, T, h = 2**(-7)):
    N = round(T/h) #Tid delt på steglengde
    t = np.linspace(0, T, N+1)
    x = np.zeros(N+1)

    # Fyller arrayet med funskjonsverdier
    x[0] = x_0 
    for k in range(N):
        x[k+1] = x[k] + h*f(x[k],k)

    plt.plot(t, x)
```

```python
def f(x, k):
    return 1   

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(x, k):
    return k  

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(x, k):
    return 4 + 3*x  

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(x, k):
    return -4*x-3*k 

diff(f, 0, 1, h = 2**(-7))
```

```python
def f(x, k):
    if k == 0:
        return 0
    
    else:
        return (3*k**2 + 4*k - x)/k 

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(x, k):
    return k + x**2

diff(f, 0, 1, h = 2**(-4))
```
