Muffins 

```python
import numpy as np

g = 9.81
kk = 0.2
m = 0.5 
h = 2**(-6)
T = 2
N = round(T/h)
t = np.linspace(0, T, N+1)
a = np.zeros(N+1)
v = np.zeros(N+1)
s = np.zeros(N+1)
```

```python
for k in range(N):
    a[k] = g - (kk/m)*v[k]**2
    v[k+1] = v[k] + h*a[k]
    s[k+1] = s[k] + h*v[k]
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.figure(figsize=(10,3))
plt.plot(t, a)
plt.show()

plt.figure(figsize=(10,2))
plt.plot(t, v, 'r')
plt.show()

plt.figure(figsize=(10,2))
plt.plot(t, s, 'g')
plt.show()
```
