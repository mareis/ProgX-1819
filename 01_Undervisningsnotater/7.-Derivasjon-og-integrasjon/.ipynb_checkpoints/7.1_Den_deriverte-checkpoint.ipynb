# 7.1 Derivasjon

## Definisjonen av den deriverte

Den deriverte til en funksjon $f$ i et punkt $a$ er stigningstallet mellom a og et punkt som er uendelig nærme a. Vi kaller dette punktet $a+\Delta x$, og den deriverte er gitt ved:

$$f^\prime(a)=\lim_{\Delta x \to 0} \frac{f(a+\Delta x) - f(a)}{\Delta x}$$

## Newtons kvotient
Vi tilnærmer den deriverte med Newtons kvotient. Den deriverte til en funksjon $f$ kan tilnærmes slik dersom $h$ er liten.

$$f^\prime(a)=\lim_{\Delta x \to 0} \frac{f(a + \Delta x) - f(a)}{\Delta x} \approx \frac{f(a + h) - f(a)}{h}$$


## Implimentering i Python

```python
def derivert(f, a, h=10**-8):
    return (f(a + h) - f(a))/h
```

### Eksempel 1

```python
def f(x):
    return x**2

print(f'{derivert(f,2):.7f}')
```

### Eksempel 2

$$g(x)=x^4 - 1$$

$$g^\prime (3)= 4 \cdot 3^3=108$$

Plot diferansen mellom den eksakt deriverte til $g^\prime (3)$ og den tilnærmede numeriske med ulike $h$.


```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def g(x):
    return x**4 - 1

avvikene = []
h = []
for i in range(3, 14):
    deriverte = derivert(g, 3, 10**-i)
    avvik = np.abs(deriverte - 108)
    avvikene.append(avvik)
    h.append(f'10e-{i}')
   
plt.figure(figsize=(10,4))
plt.xlabel('h') 
plt.ylabel('$g^{\prime}(3)$')
plt.title('Analytisk vs numerisk ved ulike h') 
plt.plot(h, avvikene)
plt.show()
```

### Lars sitt 3D-plot

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
curve_t = t
curve_y = y(t)
curve_x = x(t)

curve_dy = derivert(y, t)
curve_dx = derivert(x, t)


fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(curve_t, curve_y, curve_x)
ax.plot(curve_t, curve_dy, curve_dx)
ax.set_xlabel('t')
ax.set_ylabel('east')
ax.set_zlabel('north')
plt.legend(['Position', 'Velocity'])
plt.show()
```

### Oppgave 7.10

```python
import numpy as np
```

```python
def u(x):
    return x**3 + 3*x
```

```python
def v(x):
    return np.sqrt(np.log(x) + x**2)
```

```python
def uv(x):
    return u(x)*v(x)
```

```python
venstre1 = derivert((uv),7)
hoyre1 = derivert(u,7)*v(7) + derivert(v,7)*u(7)
```

```python
print(f'{venstre1:.2f} = {hoyre1:.2f}')
```

```python
def udelv(x):
    return u(x)/v(x) 
```

```python
venstre2 = derivert(udelv, 7)
hoyre2 = (derivert(u,7)*v(7) - derivert(v,7)*u(7))/(v(7)**2)
```

```python
print(f'{venstre2:.2f} = {hoyre2:.2f}')
```
