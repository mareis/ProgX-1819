**a)** Lag et program som generer $N = 1000$ tilfeldige verdier ved bruk av uniform.

$$m = \frac{\text{sum over alle } θ\text{-verdier som gjorde at kanonen traff blinken}} {\text{antall } θ\text{-verdier som gjorde at kanonen traff blinken}}$$

````python
liste_treff_uniform = []

def hoyde_kanonkule(g, v, theta):
    theta = np.radians(theta)
    return -50 * g/(v**2*np.cos(theta)) + 10 * np.tan(theta)

for x in range(1, n+1):
    theta = rd.uniform(30, 61)
    h = hoyde_kanonkule(g, v, theta);
    if h >= 5 and h <= 5.25:
        liste_treff_uniform.append(theta)
        
m = sum(liste_treff_uniform) / len(liste_treff_uniform)
print(f'm = {m:.2f}')
len(liste_treff_uniform)
````

### Oppgave 6.10
En måte en kan tilnærme $π$ på er ved å beregne arealet til en sirkel med radius r=1. For å tilnærme oss området til sirkelen lager vi et kvadrat slik at sirkelen passer akkurat inn i kvadratet. Kvadratet vil ha en sidelengde på 2r = 2. Deretter tegner vi N tilfeldige punkter i kvadratet og teller antall punkter M som befinner seg innenfor eller på sirkelen.

En kan finne arelaet til en sirkel med radius r ved å bruke formelen $A = πr^2$. Er $r = 1$, vil arealet være lik $π$.

<img src='pi.png' width='400px'>

Figur: Sirkel med radius lik 1 innskrevet i et kvadrat med sidelengde 2. Her er det 1000 tilfeldig genererte punkter.


Forholdet $\frac{M}{N}$ vil da fortelle oss hvor stort område sirkelen dekker i forhold til kvadratet.
Vi kan derfor finne hvor stort arealet til sirkelen er ved: 

$$\text{omtrentlig areal til sirkel }= \frac{M}{N} · A$$

der A er arealet til kvadratet.

Lag et program som generer N tilfeldige punkter (du kan bestemme en verdi for N, eller la brukeren taste inn verdi for N). Programmet skal generere tilfeldige verdier til punktenes x- og y-koordinater og angi en tilnærma verdi for pi. Et punkt med koordinater $(x, y)$ vil befinne seg inni sirkelen dersom

$$x^2 + y^2 \leq 1$$



#### Løsning:

````python
n = 1000000
x = 2 * np.random.rand(n)-1
y = 2 * np.random.rand(n)-1

m = 0
for i in range(n):
    if x[i]**2 + y[i]**2 <= 1:
        m += 1
print(f'Tilnærmingsverdi for pi: {4*m/n:.3f}')
````

### Oppgave 6.11
I denne oppgava får du se hvordan en kan lage en modell for prisutviklingen til en aksje.

**a)** Prisutviklingen per aksje hos selskapet Epler&Pærer AS følger modellen:

$$y_i =
  \begin{cases}
    y_{i-1}+1,  & \quad p_i\geq \frac{1}{2}\\
    y_{i-1}-1,  & \quad \text{ellers}
  \end{cases}, \text{ for } i=1,2,...,N-1,N 
$$


der $p_i$ er en tilfeldig verdi mellom 0 og 1 som genereres for hver i og $y_i$ er hvor mye en aksje synker/øker på i-te dag. Aksjeprisen synker dersom yi er negativ og øker dersom yi er positiv. Vi antar børsen holder åpent hver dag og lar $y_0 = 0$. Lag et program som modeller prisutviklingen til aksjene hos Epler&Pærer AS over $N = 300$ dager. Plot resultatet.

#### Løsning:

````python
n = 300
y = np.zeros(n)
p = np.random.rand(n)
for i in range(1, n):
    if p[i] >= 0.5:
        y[i] = y[i-1] + 1 
    else:
        y[i] = y[i-1] - 1
    
y
````


**b)** Prisutviklingen per aksje hos selskapet NoensBurger AS følger en litt annerledes
modell:

$$z_i =
  \begin{cases}
    z_{i-1}+1,  & \quad p_i\geq p_{i-1}\\
    z_{i-1}-1,  & \quad \text{ellers}
  \end{cases}, \text{ for } i=1,2,...,N-1,N 
$$
der også $z_0 = 0$.

Utvid programmet ditt fra a) slik at det også modellerer prisutviklingen til aksjene hos Noens Burger AS, også over N = 300 dager.
#### Løsning:

````python
z = np.zeros(n)
p = np.random.rand(n)
for i in range(1, n):
    if p[i] >= p[i-1]:
        z[i] = z[i-1] + 1 
    else:
        z[i] = z[i-1] - 1
    
y
````

**c)** Utvid programmet ditt slik at det plotter prisutviklingen til aksjene hos selska- pene i samme plott. Bruk legend (merkelapp for hver graf ) til å skille mellom grafene. Kjør programmet ditt et par ganger. Kommenter forskjellen i prisutviklingen hos Noens Burger AS og Epler&Pærer AS.

#### Løsning:

````python
x = range(n)
plt.plot(x, y, label='Epler&Pærer AS')
plt.plot(x, z, label='NoensBurger AS')
plt.legend()
plt.title('Aksjeutvikling', fontsize=16)
plt.xlabel('Tid (dager)')
plt.ylabel('Verdi ( Kr )')
plt.show()
````

**a)** Lag et program som kaster to terninger 10 000 ganger. La programmet lagre hvor mange ganger ulike par på formen (antall øyne på første terning, antall øyne på andre terning) forekommer. Dette kan for eksempel gjøres ved å ge- nerere en matrise som er 6 × 6 stor. Da vil verdien på plassen matrise[antall øyne på første terning–1][antall øyne på andre terning–1] representere antall ganger vi har fått tilhørende par.

<img src='terninger.png' width='700px'>


````python
terningmatrise = np.zeros((6, 6))
n = 10000
for x in range(n):
    terning_1 = np.random.randint(1,7) 
    terning_2 = np.random.randint(1,7)
    terningmatrise[terning_1 - 1, terning_2 - 1] += 1
terningmatrise
````
