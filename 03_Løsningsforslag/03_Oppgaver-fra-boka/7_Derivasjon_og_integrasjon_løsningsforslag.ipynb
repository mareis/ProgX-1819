### Oppgave 7.1
**Newtons kvotient:**
Skriv en funksjon *derivert* slik at delta_x har en standardverdi. 

#### Løsning:

```python
def derivert(f, a, delta_x=1e-08):
    '''
        Deriverte med Newtons kvotient
        Input: f: funksjonen, x: x-verdi, delta_x: punktavstand
    '''
    return ( f(a+delta_x) -  f(a) )/delta_x 
```

### Oppgave 7.2

```python
def utskrift(f, navn):
    print(f'{navn}´(2) = {derivert(f,2):.1f}')
    print(f'{navn}´(3) = {derivert(f,3):.1f}')
    print(f'{navn}´(4) = {derivert(f,4):.1f}')
```

```python
def graf(f, navn):
    x = np.linspace(2,5,100) # 100 x-verdier fra 2 til 5
    f_verdier = np.zeros(len(x))
    f_deriverte_verdier = np.zeros(len(x))
    for i in range(len(x)):
        f_verdier[i] = f(x[i])
        f_deriverte_verdier[i] = derivert(f, x[i])
    plt.rcParams['figure.figsize'] = (12, 4.5)
    plt.plot(x, f_verdier, label=f'{navn}')
    plt.plot(x, f_deriverte_verdier, label=f'{navn}-derivert')
    plt.legend()
    plt.title(f'{navn} og dens deriverte', fontsize=16)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() 
```

### Oppgave 7.3 (Biologi)

En plantepopulasjon følger denne utviklingen:

$$N(t)=0.2\cdot e^t + 90, \quad t \in [0,10]$$

der $N$ er antall planter ved tida t i år og $e$ er eulertallet.
Lag et program som plotter både antall planter i økosystemet $N(t)$ og endringen i antall planter $N'(t)$ i samme plot. Sett merkelapper på de to funksjonene ved å bruke legend.

#### Løsning:

```python
def N(t):
    return 0.2*np.e**t + 90

t = np.linspace(0,8,1000)

plt.plot(t, N(t), label='$N(t)$')
plt.plot(t, derivert(N,t), label='$N^\prime (t)$')
plt.legend()
plt.title('Plantepopulasjon', fontsize=16)
plt.xlabel('t (år)')
plt.ylabel('Antall planter')
plt.show() 
```

### Oppgave 7.4

En partikkel følger posisjonsfunksjonen $x(t) = t^3 + \frac13t$. Plot posisjon, hastighet og akselerasjon i hvert sitt underplott for $t \in [0,10].$

#### Løsning:

### Oppgave 7.5

```python
def x(t):
    return 3*t**2 + 4*t

def y(t):
    return 5*t + 4

t = np.linspace(0,120,1000)

plt.plot(x(t), y(t))
plt.show()
```

```python
plt.plot(t, derivert(x,t))
plt.plot(t, derivert(y,t))
```

```python
from mpl_toolkits.mplot3d import Axes3D

curve_t = t
curve_y = y(t)
curve_x = x(t)

curve_dy = derivert(y, t)
curve_dx = derivert(x, t)


fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(curve_t, curve_y, curve_x)
ax.plot(curve_t, curve_dy, curve_dx)
ax.set_xlabel('t')
ax.set_ylabel('east')
ax.set_zlabel('north')
plt.legend(['Position', 'Velocity'])
plt.show()
```

### Oppgave 7.10

```python
import numpy as np

def u(x):
    return x**3 + 3*x

def v(x):
    return np.sqrt(np.log(x) + x**2)

def uv(x):
    return u(x)*v(x)

venstre1 = derivert((uv),7)
hoyre1 = derivert(u,7)*v(7) + derivert(v,7)*u(7)


print(f'{venstre1:.2f} = {hoyre1:.2f}')
```

```python
def udelv(x):
    return u(x)/v(x) 

venstre2 = derivert(udelv, 7)
hoyre2 = (derivert(u,7)*v(7) - derivert(v,7)*u(7))/(v(7)**2)

print(f'{venstre2:.2f} = {hoyre2:.2f}')
```
