# 8. Differensialligninger - Løsningsforslag

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'

def diff(f, y_0, T, h = 2**(-7)):
    N = round(T/h) #Tid delt på steglengde
    t = np.linspace(0, T, N+1)
    y = np.zeros(N+1)

    # Fyller arrayet med funskjonsverdier
    y[0] = y_0 
    for k in range(N):
        y[k+1] = y[k] + h*f(y[k],k)

    plt.rcParams['figure.figsize'] = (12, 4.5)    
    plt.plot(t, y)
```

```python
def f(y, k):
    return 1   

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(y, k):
    return k  

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(y, k):
    return -4*y - 3*k 

diff(f, 0, 1, h = 2**(-7))
```

```python
def f(y, k):
    if k == 0:
        return 0
    
    else:
        return (3*k**2 + 4*k - y)/k 

diff(f, 0, 5, h = 2**(-7))
```

```python
def f(y, k):
    return k + y**2

diff(f, 0, 1, h = 2**(-4))
```
