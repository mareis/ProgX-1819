### Oppgave 5.9
Lag et program som sjekker om ei gitt liste inneholder et partall. Dersom lista gjør det, slett elementene som er partall.
#### Løsning (Metode 1 med liste, while-løkke og pop)

```python
liste_tilfeldig = [np.random.randint(1000) for x in range(11)] 
print(f' Liste med tilfeldige verdier: {liste_tilfeldig}')

i = 0
while i < len(liste_tilfeldig):
    if liste_tilfeldig[i] % 2 == 0:
        liste_tilfeldig.pop(i)
    else: i += 1

print(f' Liste med kun oddetall: {liste_tilfeldig}')
```

#### Løsning (Metode 2 med liste der vi lager ny liste til oddetallene)

```python
liste_tilfeldig = [np.random.randint(1000) for x in range(11)] 
print(f' Liste med tilfeldige verdier: {liste_tilfeldig}')

liste_oddetall = [x for x in liste_tilfeldig if x % 2 != 0]

print(f' Liste med kun oddetall: {liste_oddetall}')
```

### Oppgave 5.29
Dersom en person tar medisin, vil konsentrasjonen av medisinen i blodet avta over tid. En modell for hvor mye konsentrasjon $K(t)$ av medisinen som vil være igjen etter $t$ timer, er

$$K(t)=K_0e^{-at}$$

der $K_0$ er konsentrasjonen av medisinen personen har i blodet med en gang etter å ha tatt medisinen i mg/l. Verdien til a forteller oss hvor mye av medisinen som avtar per time. Lag et program som plotter $K(t)$ for $K_0 = 4.73$mg/l og $a = 0.65$ i løpet av $48$ timer. Husk å ha med tittel og enheter på aksene.

### Oppgave 5.30 (Fysikk)
Når et legeme er i fritt fall, vil den mekaniske energien være bevart, det vil si at verdien til energien er den samme for alle tider t. Den mekaniske energien E(t) ved ei tid t er gitt som

$$E(t)=E_K(t)+E_p(t)$$

der $E_k(t)=\frac{1}{2}m \cdot v(t)^2$ er den kinetiske energien ved ei tid $t$ og $E_p(t)=m\cdot g\cdot y(t)$ er den potensielle energien ved ei tid $t$. Dersom et legeme slippes fra en klippe som er $33$ meter høy, kan høyden $y(t)$ og farten $v(t)$ uttrykkes som:

$$y(t)=-\frac12 \cdot g \cdot t^2 +33$$

$$v(t)=-g\cdot t$$

Må gjøres fedig



```python
g = 9.81
m = 2.75
t = np.linspace(0,np.sqrt(66/g), 100)
y = -(1/2) * g * t**2 +33
v = -g*t
E_k = (1/2) * m * v**2
E_p = m * g * y 
    
plt.plot(t, E_k, label='Kinetisk energi')
plt.plot(t, E_p, label='Potensiell energi')
plt.plot(t, E_k+E_p, label='Mekanisk energi')
plt.legend()
plt.title('Energi til legeme i fritt fall', fontsize=16)
plt.xlabel('Tid (sek)')
plt.ylabel('Energi ( $J$ )')
plt.show()
```

### 5.31 Den generelle sinusfunksjonen: $f(x) = A \cdot \sin(kx+\phi) + d$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
x = np.linspace(0, np.pi, 100)

A = 1
k = 1
phi = 1
d = 1
f_1 = A * np.sin(k*x + phi) + d
phi = 3
f_2 = A * np.sin(k*x + phi) + d
d = 2
k = 2
f_3 = A * np.sin(k*x + phi) + d

plt.figure(1)
plt.subplot(211)
plt.plot(x, f_1)
plt.title('$A = 1, k = 1, \phi = 1, d = 1$', fontsize=16)

plt.figure(2)
plt.subplot(211)
plt.plot(x, f_2)
plt.title('$A = 1, k = 1, \phi = 3, d = 1$', fontsize=16)

plt.figure(3)
plt.subplot(211)
plt.plot(x, f_3)
plt.title('$A = 1, k = 2, \phi = 3, d = 2$', fontsize=16)


plt.show()
```
